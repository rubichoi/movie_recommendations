*   v is the number of votes for the movie
*   m is the minimum votes required to be listed in the chart (충분히 평가가 된 영화만 뽑겟다, 평가수가 상위 10%인 것을 뽑겟다)
*   R is the average rating of the movie
*   C is the mean vote across the whole report 





```python
C = df2['vote_average'].mean()
C
```

```python
m = df2['vote_count'].quantile(0.9) # (충분히 평가가 된 영화만 뽑겟다, 평가수가 상위 10%인 것을 뽑겟다)
m
```

```python
q_movies = df2.copy().loc[df2['vote_count'] >= m]
q_movies.shape
```

```python
q_movies['vote_count'].sort_values()
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v + m) * R) + (m / (m + v) * C)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1) # axis =1은 row단위로 스코어 계산산
q_movies.head(3)
```

```python
q_movies = q_movies.sort_values('score', ascending=False) #내림차순정렬렬
q_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
# 그 영화가가 우리 전체 데이터 중에서 몇번째에 해당하는지 알아보자자 인덱스가 몇번째인지.
indices = pd.Series(df2.index, index=df2['title']).drop_duplicates() # 인덱스 시리즈 만들기기
indices
```

```python
indices['The Dark Knight Rises']
```

```python
# 영화의 제목을 입력받으면 코사인 유사도를 통해서 가장 유사도가 높은 상위 10개의 영화 목록 반환
def get_recommendations(title, cosine_sim=cosine_sim):
    # 영화 제목을 통해서 전체 데이터 기준 그 영화의 index 값을 얻기
    idx = indices[title]
    
    # 코사인 유사도 매트릭스 (cosine_sim) 에서 idx 에 해당하는 데이터를 (idx, 유사도) 형태로 얻기
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # 코사인 유사도 기준으로 내림차순 정렬
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # 자기 자신을 제외한 10개의 추천 영화를 슬라이싱
    sim_scores = sim_scores[1:11]
    
    # 추천 영화 목록 10개의 인덱스 정보 추출
    movie_indices = [i[0] for i in sim_scores]
    
    # 인덱스 정보를 통해 영화 제목 추출
    return df2['title'].iloc[movie_indices]
```

```python
# 감독 정보를 추출
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
# 처음 3개의 데이터 중에서 name 에 해당하는 value 만 추출
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    return []
```

```python
features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
# 벡터화할때 띄어쓰기 반영해서 제임스따로로 카메론 따로 인식되니까 모든 단어 문자를 소문자, 띄어쓰기 없애기기
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(' ', '')) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(' ', ''))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

- '구분자'.join(리스트)
'구분자'.join(리스트)를 이용하면 리스트의 값과 값 사이에 '구분자'에 들어온 구분자를 넣어서 하나의 문자열로 합쳐줍니다.

- '_'.join(['a', 'b', 'c']) 라 하면 "a_b_c" 와 같은 형태로 문자열을 만들어서 반환해 줍니다.

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
df2['soup']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])
count_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
cosine_sim2
```
